```python
import jax
import jax.numpy as jnp
import numpy as np

jax.config.update("jax_enable_x64", True)
rng = np.random.default_rng(seed=0)
```

```python
x = rng.random((64, 64))
```

```python
x_jax = jnp.array(x, dtype=jnp.bfloat16)
```

```python
o = jax.lax.acos(x_jax)
o
```

```python
o = np.array(o, dtype=np.float64)
```

```python
np.savez_compressed("test_tpu.npz", x=x, o=o)
```
