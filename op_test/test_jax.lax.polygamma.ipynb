### 测试 jax.lax.round 算子

```python
import torch
import jax.numpy as jnp
import numpy as np
import jax
```

```python
data = np.load("/Users/haifeng/Program/FUEL-JAX/input/jax.lax.polygamma/00.npz")
m = data["m"]
x = data["x"]
print(m)
x
```

```python
j_m = m.astype(jnp.float32)
j_x = jnp.array(x, dtype=jnp.float32)
t_m = int(m)
t_x = torch.from_numpy(x).to(torch.float32)
```

```python
j_out = jax.lax.polygamma(j_m, j_x)
j_out
```

```python
print(jax.scipy.special.polygamma(int(j_m), j_x))
print(torch.special.polygamma(t_m, t_x))
```

```python
# 输入 x < 0
print(f"input:{x[0][1]}")
print(f"jax_out:{j_out[0][1]}")
print(f"torch_out:{t_out[0][1].numpy()}")
```

## 说明，这两个函数在 输入是正数的时候是一致的

```python
x = np.abs(x) + 0.001
j_x = jnp.array(x, dtype=jnp.float32)
t_x = torch.from_numpy(x).to(torch.float32)
```

```python
x
```

```python
j_out = jax.lax.polygamma(j_m, j_x)
j_out
```

```python
t_out = torch.ops.aten.special_polygamma.default(t_m, t_x)
t_out
```

```python
j_out_np = np.array(j_out, dtype=np.float64)
j_out_np
```

```python
t_out_np = t_out.detach().cpu().to(torch.float64).numpy()
t_out_np
```

```python
diff = np.abs(j_out_np - t_out_np)
diff
```
