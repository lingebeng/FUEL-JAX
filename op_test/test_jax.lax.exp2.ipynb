### 测试 jax.lax.exp2 算子

```python
import torch
import jax.numpy as jnp
import numpy as np
import jax
```

```python
# data = np.load("/Users/haifeng/Program/FUEL-JAX/input/jax.lax.exp2/00.npz")
# x = data['x']
# x
rng = np.random.default_rng(0)
x = rng.normal(size=(1024, 1024)).astype(np.float32)
```

```python
j_x = jnp.array(x, dtype=jnp.bfloat16)
t_x = torch.from_numpy(x).to(torch.bfloat16)
```

```python
j_out = jax.lax.exp2(j_x)
j_out
```

```python
t_out = torch.ops.aten.exp2.default(t_x)
t_out
```

```python
j_out_np = np.array(j_out, dtype=np.float32)
t_out_np = t_out.detach().to("cpu").to(torch.float32).numpy()
```

```python
j_out_np
```

```python
t_out_np
```

```python
diff = np.abs(j_out_np - t_out_np)
max_diff = np.nanmax(diff) if not np.isnan(diff).all() else 0.0
max_diff
```
