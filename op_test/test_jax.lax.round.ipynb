### 测试 jax.lax.round 算子

```python
import torch
import jax.numpy as jnp
import numpy as np
import jax
```

```python
# data = np.load("/Users/haifeng/Program/FUEL-JAX/input/jax.lax.round/00.npz")
# x = data['x']
# x
x = np.array([1.5, 2.5, 2.51, 3.5])
```

```python
j_x = jnp.array(x, dtype=jnp.bfloat16)
t_x = torch.from_numpy(x).to(torch.bfloat16)
```

```python
j_out = jax.lax.round(j_x)
j_out
```
