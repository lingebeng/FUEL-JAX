```python
import torch
import numpy as np


def get_ulp_diff_matrix_float(pred, truth):
    """
    计算 ULP 差异矩阵 (浮点数版本)。
    表示 pred 和 truth 之间相差了多少个机器精度单位。
    """
    # 1. 确保转为 Numpy 且为 float32
    a = np.array(pred, dtype=np.float32)
    b = np.array(truth, dtype=np.float32)

    # 2. 计算绝对误差
    abs_diff = np.abs(a - b)

    # 3. 计算真值处的 1 ULP 长度 (Gap size)
    # np.spacing(x) 返回 x 到下一个可表示浮点数的距离
    gap = np.spacing(b)

    # 4. 处理除零风险 (当 b=0 时，gap=0)
    # 0 处的 ULP 通常定义为最小正非零数 (subnormal min) 或 min normal
    # 这里我们用 float32 的最小正数填充，避免 NaN
    min_val = np.finfo(np.float32).tiny  # 约 1.17e-38
    gap[gap == 0] = min_val

    # 5. 计算比率
    ulp_diff = abs_diff / gap

    return ulp_diff


# === 使用示例 ===
# j_out_torch 和 t_out 是你的张量
# ulp_matrix = get_ulp_diff_matrix_float(j_out_torch, t_out)

# print("最大 ULP 误差:", np.max(ulp_matrix))
# print("平均 ULP 误差:", np.mean(ulp_matrix))
```
